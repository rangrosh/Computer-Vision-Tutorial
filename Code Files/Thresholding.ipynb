```python
import cv2
import matplotlib.pyplot as plt
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "gray21.512.tiff"
img2p = path + "4.2.07.tiff"
img1 = cv2.imread(img1p)
img2 = cv2.imread(img2p)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

th = 127 
max_val = 255
titles = ['img1', 'binary', 'binary inverted', 'zero', 'zero inv', 'trunc']
titles2 = ['img2', 'binary', 'binary inverted', 'zero', 'zero inv', 'trunc']

ret, output1 = cv2.threshold(img1, th, max_val, cv2.THRESH_BINARY) #if value greater than th then set to max, else 0
ret, output2 = cv2.threshold(img1, th, max_val, cv2.THRESH_BINARY_INV) # opposite of thresh
ret, output3 = cv2.threshold(img1, th, max_val, cv2.THRESH_TOZERO) # if value less than th then set to 0
ret, output4 = cv2.threshold(img1, th, max_val, cv2.THRESH_TOZERO_INV) #if value greater than th then set to 0
ret, output5 = cv2.threshold(img1, th, max_val, cv2.THRESH_TRUNC) #if value greater than th then set to value at th
images = [img1, output1, output2, output3, output4, output5]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()


#Color image
ret, output1 = cv2.threshold(img2, th, max_val, cv2.THRESH_BINARY)
ret, output2 = cv2.threshold(img2, th, max_val, cv2.THRESH_BINARY_INV)
ret, output3 = cv2.threshold(img2, th, max_val, cv2.THRESH_TOZERO)
ret, output4 = cv2.threshold(img2, th, max_val, cv2.THRESH_TOZERO_INV)
ret, output5 = cv2.threshold(img2, th, max_val, cv2.THRESH_TRUNC)
images = [img2, output1, output2, output3, output4, output5]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles2[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

# Otsu's Binarization Thresholding

```python
#Otsu's helps seperate the foreground with the background during thresholding

import cv2
import matplotlib.pyplot as plt
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "cameraman.tif"
img1 = cv2.imread(img1p , 0)

th = 0 #For otsu's you just set th = 0
max_val = 255
titles = ['img1', 'binary', 'binary inverted', 'zero', 'zero inv', 'trunc']

ret, output1 = cv2.threshold(img1, th, max_val, cv2.THRESH_BINARY + cv2.THRESH_OTSU) #if value greater than th then set to max, else 0
ret, output2 = cv2.threshold(img1, th, max_val, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU) # opposite of thresh
ret, output3 = cv2.threshold(img1, th, max_val, cv2.THRESH_TOZERO+ cv2.THRESH_OTSU) # if value less than th then set to 0
ret, output4 = cv2.threshold(img1, th, max_val, cv2.THRESH_TOZERO_INV + cv2.THRESH_OTSU) #if value greater than th then set to 0
ret, output5 = cv2.threshold(img1, th, max_val, cv2.THRESH_TRUNC + cv2.THRESH_OTSU) #if value greater than th then set to value at th
images = [img1, output1, output2, output3, output4, output5]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i] , cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

# Adaptive Thresholding

```python
#If an image has only two components a foreground and a background then OTSU is a good choice
#If there are different lighting conditions, then adaptive is preffered

import cv2
import matplotlib.pyplot as plt
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "5.1.12.tiff"
img1 = cv2.imread(img1p , 0)

block_size = 11 #has to be odd #number of blocks around the point considered
constant = 7
mean = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, block_size, constant)
gaussian = cv2.adaptiveThreshold(img1, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)

titles = ['img1', 'Mean adaptive', 'Gaussian adaptive']
images = [img1, mean, gaussian]

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i] , cmap = 'gray')
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
