```python
#Image Quantization means only using a few colours to view the image
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.2.07.tiff"
img1 = cv2.imread(img1p)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
Z = img1.reshape((-1,3))
Z = np.float32(Z) #converting from uint8 to float32

criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10,1.0)

K = 2 #No of colours
ret, label1, center1 = cv2.kmeans(Z,K,None,criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center1 = np.uint8(center1)
res1 = center1[label1.flatten()]
output1 = res1.reshape((img1.shape))

K = 8
ret, label1, center1 = cv2.kmeans(Z,K,None,criteria, 10, cv2.KMEANS_RANDOM_CENTERS)
center1 = np.uint8(center1)
res1 = center1[label1.flatten()]
output2 = res1.reshape((img1.shape))

images = [img1, output1, output2]
for i in range(3):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i])
    
plt.show()
```
