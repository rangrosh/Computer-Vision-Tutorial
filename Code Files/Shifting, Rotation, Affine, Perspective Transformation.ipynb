## Rotation

```python
#We use cv2.getRotationMatrix2D to get transformation Matrix

import cv2
import matplotlib.pyplot as plt
import numpy as np
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img1 = cv2.imread(img1p)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

rows, columns, channels = img1.shape
R = cv2.getRotationMatrix2D((columns/2,rows/2), 30 , 0.5) # origin of image(centre), rotation anticlockwise, Scaling(in both axes)
print(R)
output = cv2.warpAffine(img1, R, (columns, rows))

plt.imshow(output)
plt.show()
```

# Affine Transformation

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img1 = cv2.imread(img1p)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

pts1 = np.float32([[50,50],[200,50],[50,200]]) #these are 3 non colinear points
pts2 = np.float32([[10,100],[200,50],[100,250]]) #position of 3 points after transformation
M = cv2.getAffineTransform(pts1,pts2)

rows, columns, channels = img1.shape
output = cv2.warpAffine(img1, M, (columns, rows))

plt.imshow(output)
plt.show()
```

# Perspective Transformation

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img1 = cv2.imread(img1p)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

pts1 = np.float32([[50,50],[200,50],[50,200],[100,200]]) #these are 4 non colinear points
pts2 = np.float32([[100,100],[400,100],[100,400], [200,400]]) #position of 4 points after transformation
M = cv2.getPerspectiveTransform(pts1,pts2)

rows, columns, channels = img1.shape
output = cv2.warpPerspective(img1, M, (300, 300)) #img, M, size of output image, 700,700 will zoom out

plt.imshow(output)
plt.show()
```
