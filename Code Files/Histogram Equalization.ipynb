```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.2.07.tiff"
img1 = cv2.imread(img1p)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

r,g,b = cv2.split(img1)
output1_r = cv2.equalizeHist(r)
output1_g = cv2.equalizeHist(g) 
output1_b = cv2.equalizeHist(b) 
output1 = cv2.merge((output1_r,output1_g,output1_b))

clahe = cv2.createCLAHE(clipLimit = 2.0, tileGridSize = (8,8))
output2_r = clahe.apply(r)
output2_g = clahe.apply(g) 
output2_b = clahe.apply(b) 
output2 = cv2.merge((output2_r,output2_g,output2_b))

images = [img1, output1, output2]
for i in range(3):
    plt.subplot(3,1,i+1)
    plt.imshow(images[i])
    plt.xticks([])
    plt.yticks([]) 
    

    
plt.show()
```
