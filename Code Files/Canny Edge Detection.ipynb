```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "5.1.11.tiff"
img1 = cv2.imread(img1p,0)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

L1 = cv2.Canny(img1, 80,150, L2gradient = False)
L2 = cv2.Canny(img1, 80,150, L2gradient = True)

images = [img1, L1, L2]
for i in range(3):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i], cmap = 'gray')
    
plt.show()
```
