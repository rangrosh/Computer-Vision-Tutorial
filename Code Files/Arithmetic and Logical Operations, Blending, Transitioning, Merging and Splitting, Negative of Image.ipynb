# Blending

```python
import cv2
import matplotlib.pyplot as plt
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img2p = path + "4.1.05.tiff"
img1 = cv2.imread(img1p)
img2 = cv2.imread(img2p)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

alpha = 0.5
beta = 0.5 #aplha +beta = 1
gamma = 0
blend = cv2.addWeighted(img1, alpha, img2, beta, gamma) # blend = img1*alpha + img2*beta + gamma
titles = ['img1', 'img2', 'blend']
images = [img1, img2, blend]

for i in range(3):
    plt.subplot(1,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```

# Transition

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import time
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img2p = path + "4.1.05.tiff"
img1 = cv2.imread(img1p)
img2 = cv2.imread(img2p)

for i in np.linspace(0,1,20):
    alpha = i 
    beta = 1 - alpha
    transition = cv2.addWeighted(img1, alpha, img2, beta, 0)
    cv2.imshow('Transition',transition)
    time.sleep(0.1)
    if cv2.waitKey(1) == 27:
        break
        
cv2.destroyAllWindows()
```

# Merging and Splitting

```python
import cv2
import matplotlib.pyplot as plt
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img1 = cv2.imread(img1p)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)

r,g,b = cv2.split(img1)
titles = ['img1', 'red' , 'blue', 'green']

images = [cv2.merge((r,g,b)), r , g, b]
cmapl = ['Spectral', 'Reds', 'Blues', 'Greens']

for i in range(4):
    plt.subplot(1,4,i+1)
    plt.imshow(images[i], cmap = cmapl[i])
    plt.title(titles[i])
    plt.xticks([])
    plt.yticks([])
plt.show()
```
