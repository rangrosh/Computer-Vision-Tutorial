```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = "C:\\Users\\rangr\\Desktop\\Image Processing\\misc\\"
img1p = path + "4.1.04.tiff"
img1 = cv2.imread(img1p)
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)


k = np.array(([1,0,-1], 
              [0,0,0] ,
              [-1,0,1]), np.float32)  #convolution matrix/mask for edge detection
#matrices can be found on wikipedia

output = cv2.filter2D(img1, -1, k) #-1 signifies that the depth of output and original image are same

plt.imshow(output)
 
plt.show()
```
